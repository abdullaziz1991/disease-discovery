# Disease Prediction model

# Dataset: Kaggle
https://www.kaggle.com/datasets/kaushil268/disease-prediction-using-machine-learning?resource=download&select=Training.csv

```python
import pandas as pd
#  مكتبة لتحليل البيانات وتنظيمها في جداول (DataFrames). تُستخدم بشكل واسع للتعامل مع البيانات المهيكلة. pandas

from sklearn.model_selection import train_test_split, KFold ,cross_val_score
# train_test_split: تقسيم البيانات إلى مجموعات تدريب واختبار.
# KFold: أداة لتقسيم البيانات إلى مجموعات مختلفة (تُعرف بالـ Folds) لاستخدامها في التحقق المتقاطع (Cross-Validation).
# cross_val_score: لحساب أداء النموذج باستخدام التحقق المتقاطع.

from sklearn.preprocessing import LabelEncoder
# LabelEncoder: لتحويل البيانات النصية (Categorical Data) إلى أرقام، حيث أن العديد من النماذج لا تتعامل مع النصوص.

from sklearn.metrics import classification_report, accuracy_score
# classification_report: يعطي تقريراً شاملاً عن أداء النموذج، بما في ذلك الدقة (precision)، الاستدعاء (recall)، ودرجة F1.
# accuracy_score: لحساب نسبة التوقعات الصحيحة مقارنة بعدد العينات الكلي.
```

```python
# importing dataset as a pandas dataframe and showing the first 5 records
df = pd.read_csv('diseases_dataset.csv')
df.head()
```

```python
df.describe()
# count: عدد القيم غير الفارغة (غير المفقودة).
# mean: المتوسط الحسابي.
# std: الانحراف المعياري (يقيس مدى تباعد القيم عن المتوسط).
# min: أصغر قيمة.
# 25%: الربع الأول (25% من البيانات أقل من هذه القيمة).
# 50%: الوسيط (الربع الثاني أو القيمة الوسطى).
# 75%: الربع الثالث (75% من البيانات أقل من هذه القيمة).
# max: أكبر قيمة.
```

```python

X = df.drop(columns=["prognosis"])
y = df["prognosis"]
```

```python
# encoding the labels (transforming the labels from text to numbers)
le = LabelEncoder()
y_encoded = le.fit_transform(y)
y_encoded
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
k_fold = KFold(n_splits= 5, shuffle=True, random_state=42)
```

# Multi-Layer Perceptron (MLP)

```python
from sklearn.neural_network import MLPClassifier
import joblib
```

```python
mlp_model = MLPClassifier(hidden_layer_sizes=(200, 100), activation='relu', solver='adam', max_iter=500, verbose=False)
mlp_model.fit(X_train, y_train)
```

```python

mlp_predictions = mlp_model.predict(X_test)
print("MLP Classification Report:")
print(classification_report(y_test, mlp_predictions, target_names=le.classes_))
print(f"MLP Accuracy: {accuracy_score(y_test, mlp_predictions):.2f}")
```

```python
scoring = 'accuracy'
score = cross_val_score(mlp_model, X_train, y_train, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

# prediction function

```python
joblib.dump(mlp_model, 'disease_model.pkl')
joblib.dump(le.classes_, 'label_encoder_classes.pkl')  # For decoding predictions
joblib.dump(X_test.columns.tolist(), 'feature_columns.pkl')  # Save feature order
```
