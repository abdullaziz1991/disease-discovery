```python

X = df.drop(columns=["prognosis"])
y = df["prognosis"]
```

```python
# encoding the labels (transforming the labels from text to numbers)
from sklearn.preprocessing import LabelEncoder
# LabelEncoder: لتحويل البيانات النصية (Categorical Data) إلى أرقام، حيث أن العديد من النماذج لا تتعامل مع النصوص.

le = LabelEncoder()
y_encoded = le.fit_transform(y)
y_encoded
```

```python
from sklearn.neural_network import MLPClassifier
# from sklearn.model_selection import KFold, cross_val_predict
# from sklearn.metrics import classification_report, accuracy_score
import joblib

# mlp_model = MLPClassifier(hidden_layer_sizes=(100,), max_iter=500)
mlp_model = MLPClassifier(hidden_layer_sizes=(100,), solver='adam', learning_rate_init=0.01,
                           activation='relu', max_iter=500,random_state=42)

# إعداد KFold بدون فقدان أي مرض
# k_fold = KFold(n_splits=5, shuffle=True, random_state=42)

# # تنفيذ التدريب والتنبؤ معًا لكل الطيات
# mlp_predictions = cross_val_predict(mlp_model, X, y_encoded, cv=k_fold)

# # تقرير التصنيف الكامل لجميع الأمراض
# print("MLP Classification Report:")
# print(classification_report(y_encoded, mlp_predictions, target_names=le.classes_))

# print(f"MLP Accuracy: {accuracy_score(y_encoded, mlp_predictions):.2f}")

# درّب النموذج النهائي على كل البيانات
mlp_model.fit(X, y_encoded)

# ثم احفظه
joblib.dump(mlp_model, 'disease_mlp_model.pkl')
joblib.dump(le.classes_, 'label_encoder_classes.pkl')  # أسماء الأمراض
joblib.dump(X.columns.tolist(), 'feature_columns.pkl')  # أسماء الأعراض

```

# prediction function

```python
def predict_disease(symptom_values):
    """
    Parameters:
    symptom_values (dict): Dictionary with symptom names as keys and 0/1 as values

    Returns:
    str: Predicted disease name
    """
    # Load artifacts
    model = joblib.load('disease_mlp_model.pkl')  # ✅ غيّر الاسم هنا
    feature_columns = joblib.load('feature_columns.pkl')
    le_classes = joblib.load('label_encoder_classes.pkl')

    # Create input DataFrame
    input_df = pd.DataFrame(0, index=[0], columns=feature_columns)

    # Update with input values
    for symptom, value in symptom_values.items():
        if symptom in input_df.columns:
            input_df[symptom] = value

    # Predict
    prediction = model.predict(input_df)

    # Decode label
    return le_classes[prediction[0]]
```

```python
# Example input
sample_input = {
    'high_temperature': 1,
 'cough': 1,
 'sore_throat': 1,
 'muscle_pain': 1,
 'fatigue': 1,
 'exhaustion': 1,
 'headache': 1,

    }

# Fill missing symptoms with 0
all_symptoms = joblib.load('feature_columns.pkl')
complete_input = {symptom: sample_input.get(symptom, 0) for symptom in all_symptoms}

# Get prediction
predicted_disease = predict_disease(complete_input)
print(f"Predicted Disease: {predicted_disease}")
```
